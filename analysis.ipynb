```python
# Ensure date, start and end columns are in datetime format
feeding_data['date'] = pd.to_datetime(feeding_data['date'])
feeding_data['start'] = pd.to_datetime(feeding_data['start'])
feeding_data['end'] = pd.to_datetime(feeding_data['end'])
```

```python
# Create stations dictionary (contains the pig index and the pig number for each station)
stations = {}

# Group the data by station
grouped_data = feeding_data.groupby('station')

# Fill stations dictionary
for station, group in grouped_data:
    subdictionary = dict(zip(group['pig_index'], group['pig']))
    stations[station - 1] = subdictionary
```

```python
# Filter the data for a specific date
filtered_data = feeding_data[feeding_data['date'] == '2021-03-07']

# Create dictionaries for this specific date
intake = {}
relations = {}

# Group the filtered data by station
grouped_data = filtered_data.groupby('station')

# Fill intake dictionary
for station in range(10):
    intake[station] = {}
    for pig_index in range(11):
        subset = filtered_data[
            (filtered_data['station'] == station + 1) &
            (filtered_data['pig_index'] == pig_index)]
        if not subset.empty:
            pig = int(subset['pig'].values[0])
            total_intake = round(float(subset['tot_intake_day_pig'].values[0]), 2)
        else:
            pig = int(filtered_data[(filtered_data['pig_index'] == pig_index)]['pig'].values[0])
            total_intake = 0.0  
        intake[station][pig_index] = {'pig': pig, 'tot_intake_day_pig': total_intake}

# Fill relations dictionary
for station in range(10):
    relations[station] = {}
    index = 0
    for pig_index in range(11):
        for count_index in range(pig_index + 1, 11):
            total_count = filtered_data[
                (filtered_data['station'] == station + 1) &
                (filtered_data['pig_index'] == pig_index) &
                ((filtered_data['pig_before_index'] == count_index) | (filtered_data['pig_after_index'] == count_index))
            ].shape[0]
            relations[station][index] = total_count
            index += 1
```

```python
# Get the minimum and maximum values of the 'tot_intake_day_pig' column
min_intake = feeding_data['tot_intake_day_pig'].min()
max_intake = feeding_data['tot_intake_day_pig'].max()

# Print the results
print(f"Minimum value of 'tot_intake_day_pig': {min_intake}")
print(f"Maximum value of 'tot_intake_day_pig': {max_intake}")

# Initialize variables to store the minimum and maximum values
min_relation = float('inf')
max_relation = float('-inf')

# Iterate through the relations_indexed dictionary to find the min and max values
for station in relations:
    for pig_index in relations[station]:
        value = relations[station][pig_index]
        if value < min_relation:
            min_relation = value
        if value > max_relation:
            max_relation = value

# Print the minimum and maximum values
print(f"Minimum value in relations: {min_relation}")
print(f"Maximum value in relations: {max_relation}")

def scale_value(value, min_value, max_value, new_min, new_max):
    """
    Scales a value from its original range to a new range.
    
    Parameters:
    value (float): The value to be scaled.
    min_value (float): The minimum value of the original range.
    max_value (float): The maximum value of the original range.
    new_min (float): The minimum value of the new range.
    new_max (float): The maximum value of the new range.
    
    Returns:
    float: The scaled value.
    """
    return ((value - min_value) / (max_value - min_value)) * (new_min - new_max) + new_max
```
